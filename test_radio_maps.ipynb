```python
import numpy as np
import matplotlib.pyplot as plt
from radio_map_utils import shadowing_data, generate_map, column_normalization
# from generate_radio_map import generate_radio_map
```

```python
with np.load('radio_map.npz') as data:
    T, C_true, S_true, T_true = data["T"], data["C_true"], data["S_true"], data["T_true"]
```

```python
R = C_true.shape[1]
visualize_K=25
```

```python
rows = int(np.ceil(6.0/5.0))
cols = int(np.ceil((R+1)/ rows))
fig, axs = plt.subplots(rows, cols, figsize=(6*R, 16))
for rr in range(R):
    i, j = np.unravel_index(rr, (rows, cols))
    axs[i, j].plot(C_true[:,rr])

i, j = np.unravel_index(rr+1, (rows, cols))
axs[i, j].imshow(np.log10(T_true[visualize_K].squeeze()), cmap="jet")
plt.show()
```
